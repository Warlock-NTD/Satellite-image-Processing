```python
df=pd.read_excel('/content/drive/My Drive/Knowledge/Data_Science_Prj/EPA_merged_2019.xlsx')
```

Read dataframe and extract DateTime to 3 addition feature: Year, Month, Date.

```python
df['Year'] = pd.DatetimeIndex(df['Thời gian']).year
df['Month'] = pd.DatetimeIndex(df['Thời gian']).month
df['Day'] = pd.DatetimeIndex(df['Thời gian']).day
print(df)
```

**This Part using GroupBy to calculate Average weight**

Epochs 1: Try calculating average of weight groupby Day. But it's necessary to determine that it's not just groupby day column, it also depend on Month, Year columns.

```python
df['Year-Month-Day'] = df.apply(lambda x: '%s-%s-%s' % (x['Year'], x['Month'], x['Day']), axis = 1)
print(df)
```

Calculating Average group of Year-Month-Day

```python
Day_Series_AVE = df.groupby(['Name', 'Year-Month-Day']).mean()
print(Day_Series_AVE)
```

**Note**: Ignore the value of 3 colums : Year, Month, Day

Epochs 2: Try calculating average of weight groupby Month. But it's necessary to determine that it's not just groupby Month column, it also depend on Year .

```python
df['Year-Month'] = df.apply(lambda x: '%s-%s' % (x['Year'], x['Month']), axis = 1)
print(df)
```

Calculating Average group of Year-Month

```python
Month_Series_AVE = df.groupby(['Name', 'Year-Month']).mean()
print(Month_Series_AVE)
```

**Note**: Ignore the value of 3 colums : Year, Month, Day

Epochs 3: Try calculating average of weight groupby Year.

Calculating Average group of Year

```python
Year_Series_AVE = df.groupby(['Name', 'Year']).mean()
print(Year_Series_AVE)
```

**Note**: Ignore the value of 3 colums : Year, Month, Day
