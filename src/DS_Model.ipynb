# **Start Using Model to fill the Map With Predicting Value PM2.5**

```python
from osgeo import gdal
from osgeo import ogr
import matplotlib.pyplot as plt
```

```python
# traffic avg
ds1 = gdal.Open("/content/drive/My Drive/Data/average_average_30m.tif")
ds1_band = ds1.GetRasterBand(1)
nodataval1 = ds1_band.GetNoDataValue()
gt1 = ds1.GetGeoTransform()
ds1_arr = ds1.ReadAsArray().astype(np.float64)

# Road dense
ds4 = gdal.Open("/content/drive/My Drive/Data/road_density_hn_30m_padded.tif")
ds4_band = ds4.GetRasterBand(1)
nodataval4 = ds4_band.GetNoDataValue()
ds4_arr = ds4.ReadAsArray().astype(np.float64)
gt4 = ds4.GetGeoTransform()

# Land Cover
ds5 = gdal.Open("/content/drive/My Drive/Data/lc_2017.tif")
ds5_band = ds5.GetRasterBand(1)
nodataval5 = ds5_band.GetNoDataValue()
ds5_arr = ds5.ReadAsArray().astype(np.float64)
gt5 = ds5.GetGeoTransform()

df2 = pd.read_csv('/content/drive/My Drive/Data/VVNB_2019-2021_mean_by_month.csv')
def getVVNB(year, month):
  strquery = "year == '" + str(year) + "' and month == '" + str(year) + "-"
  if month < 10: strquery = strquery + "0" + str(month) + "'"
  else: strquery = strquery + str(month) + "'"
  altm = df2.query(strquery).iloc[0, 2]
  temp = df2.query(strquery).iloc[0, 3]
  hud = df2.query(strquery).iloc[0, 4]
  return np.array([temp, altm, hud])
```

```python
def getLc(i, j):
  if ds5_arr[i][j] == 6:  return 1
  return 0
```

```python
# Full Fill and Generate Map monthly by year.
def Renderer(year, month):
  path = "/content/drive/My Drive/Data/" + str(year) + "/PM25_" + str(year)

  if month < 10: path = path + '0' + str(month) + '01_hanoi.tif'
  else: path = path + str(month) + '01_hanoi.tif'
  ds = gdal.Open(path)
  ds_band = ds.GetRasterBand(1)
  nodataval = ds_band.GetNoDataValue()
  ds_arr = ds.ReadAsArray().astype(np.float64)
  arr = np.zeros(ds_arr.shape)
  gt = ds.GetGeoTransform()
  dz = getVVNB(year, month)

  # Fill data
  for i in range(0, 3029):
    for j in range(0, 2540):
      if ds_arr[i][j] == nodataval or ds1_arr[i][j] == nodataval1 or ds5_arr[i][j] == nodataval5 or ds4_arr[i][j] == nodataval4:
        arr[i][j] = -9999
        continue
      
      input = np.array([ds_arr[i][j], ds1_arr[i][j] , getLc(i, j), ds4_arr[i][j], dz[0], dz[1], dz[2]])
      arr[i][j] = result3.predict([input])
      # print(input)
      # print(arr[i][j])
  print(arr)

  # Map create
  drv = gdal.GetDriverByName("GTiff")
  dst = drv.Create("/content/drive/My Drive/Results/Result_PM25_Hanoi_" + str(year) + "_" + str(month) + ".tif", ds.RasterXSize, ds.RasterYSize, 1, gdal.GDT_Float64)
  dst.SetGeoTransform(gt)
  dst.SetProjection(ds.GetProjection())
  dst.GetRasterBand(1).SetNoDataValue(-9999)
  dst.GetRasterBand(1).WriteArray(arr)
```

```python
Renderer(2019, 1)
```
